# Step 1] Load the Amazon Dataset 
- Understand the data

```python
import pandas as pd

data = pd.read_csv('Amazon_reviews_2023.csv')
df= pd.DataFrame(data)
df
```

# Step 2] Data Cleaning
- Drop null values

```python
df.dropna(subset=['rating', 'text'], inplace = True)
```

- Replace the null value with text in title column

```python
df['title'] = df['title'].fillna("No Title")
```

- Format to datetime 

```python
df['timestamp']= pd.to_datetime(df['timestamp'])
```

- Check the number of duplicate values

```python
df.duplicated().sum()
```

- Drop duplicate values

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

-  Reset index after cleaning

```python
df.reset_index(drop = True, inplace = True)
```

```python
df
```

# Step 3] - Exploratory Data Analysis(EDA) 
- Rating Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data = df, x = 'rating', color = 'lightblue', edgecolor = 'black')
plt.title('Rating Distribution')
plt.show()
```

- Verified vs. Non-Verified Purchases

```python
sns.countplot(data = df, x = 'verified_purchase', color = 'lightgreen', edgecolor = 'black')
plt.title('Verified Purchase Count')
plt.show()
```

-  Most Helpful Reviews

```python
top_helpful = df.sort_values(by = 'helpful_vote', ascending = False).head(10)
top_helpful[['title','text','helpful_vote']]
```

- Response based on review

```python
from textblob import TextBlob

df['polarity'] = df['text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
df['sentiment'] = df['polarity'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))

sns.countplot(x='sentiment', data=df)
plt.title("Sentiment Distribution")
plt.show()
```

#  Step 5] Time-Based Analysis

```python
df['year_month'] = df['timestamp'].dt.to_period('M')
monthly_trend = df.groupby('year_month').size()

monthly_trend.plot(kind='line',figsize = (12,6), title = 'Review Volume Over Time')
plt.ylabel('Number of Review')
plt.show()
```
